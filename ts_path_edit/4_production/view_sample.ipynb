```python
import pandas as pd
import numpy as np
from architector import view_structures, convert_io_molecule
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('all_combined.pkl')
indf = pd.read_pickle('to_gen_structures_parallel.pkl')
```

```python
charge_spindf = indf[['rxn_uid','total_charge','total_uhf','total_lig_charges']]
```

```python
rxn_name = [row['orig_name'].split('_')[0] for i,row in df.iterrows()]
df['rxn_name'] = rxn_name

natoms = []
for i,row in df.iterrows():
    if isinstance(row.sdf_file,str):
        nats = int(row.sdf_file.split('\n')[3].strip().split()[0])
        natoms.append(nats)
    else:
        natoms.append(0)
df['natoms'] = natoms
```

```python
df.natoms.hist(bins=np.arange(1,101,1))
plt.xlabel('number of atoms')
plt.ylabel('count')
```

```python
good_df = df[~df.sdf_file.isna()]
bad_df = df[df.sdf_file.isna()]
```

```python
# Combine dataframes to include the charges and spins relevant.
good_df['rxn_uid'] = [row['orig_name'].replace('.sdf','') for i,row in good_df.iterrows()]
fcharge_spindf = charge_spindf[charge_spindf.rxn_uid.isin(good_df.rxn_uid.values)]
merged = good_df.merge(fcharge_spindf,on='rxn_uid')
merged.to_pickle('good_charge_spin_all_combined.pkl')
```

```python
rs = 42
srow = merged.sample(1, random_state=rs).iloc[0]
mols= convert_io_molecule(srow['sdf_file'])
print('Reversed? ', ('rev' in srow['new_name']))
print('Initial\t\t\t\tFinal')
view_structures(mols)
```
